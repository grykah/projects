# Infographic

```python
# get packages
import matplotlib.pyplot as plt
from matplotlib.image import NonUniformImage
from matplotlib import cm

import numpy as np
```

## Plots

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(6, 6), constrained_layout=False)

title_obj = fig.suptitle('Regenerative Agriculture', fontsize='large') #get the title property handler
#plt.getp(title_obj)                    #print out the properties of title
plt.getp(title_obj, 'text')            #print out the 'text' property for title
plt.setp(title_obj, color='ivory')         #set the color of title to red



ax1.set_aspect(1)
ax1.text(0.5, 0.5, 'Soil is made of about \n 45% minerals \n 25% water \n 5% organic matter \n 25% air',
         rotation=0,
         horizontalalignment='center',
         verticalalignment='center',
         multialignment='center')

# Some example data to fill in other visuals
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
ax2.plot(x, y)
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')
ax5.plot(x, y**3, 'tab:orange')
ax6.plot(x, y**5, 'tab:purple')


#fig.patch.set_visible(False)

ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
ax4.axis('off')
ax5.axis('off')
ax6.axis('off')



fig.patch.set_facecolor('green')
```
