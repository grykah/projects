## Load Data and Packages

```python
# packages by section
# load data and packages
import os
import pandas as pd

# explore data values - univariate analysis
import seaborn as sns
import matplotlib.pyplot as plt

# dashboard - visual
import plotly.express as px

# dashboard - implementation
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
# set up path and load data.
# base path and path
PROJ_ROOT = os.path.join(os.pardir)
#print(os.path.abspath(PROJ_ROOT))
data_path = os.path.join(PROJ_ROOT, 'MentalHealth', 'mental_health_in_tech_survey.csv') 
print(data_path)

# load the data
data = pd.read_csv(data_path)
```

```python
data.head()
```

## Explore Data Values 
In the table above, it is already clear that a variety of response variations exist within each column and a multitude of NaN values will need to be addressed.  Let's explore the data a bit further to better understand possible issues.

### Null Values

```python
data.isnull().sum()
```

The 'state' column is missing a significant number of values, but this may be due to these responses being from a different country.  

The 'comments' column is missing the majority of the data and will not be useful.  We can drop this column.

Other missing values may be imputed.

```python
# examine 'state' NaN
data[data['state'].isna()]
```

```python
# replace NaN in state to indicate not U.S.
data['state'].fillna('Other',inplace=True)
```

```python
#drop 'comments'
data.drop(columns=['comments'], inplace = True)
```

```python
#impute data

#print(data['work_interfere'].value_counts())
data['self_employed'].fillna('No',inplace=True)
data['work_interfere'].fillna('Sometimes',inplace=True)
```

```python
data.isnull().sum()
```

### Univariate Analysis

```python
sns.boxplot(data['Age']);
```

The wide variance in the 'Age' columns indicates outliers.  It is not likely that someone under the age of 10 or over the age of 90 are working in tech.  

```python
data.drop(data[data['Age'] < 10].index, inplace = True) 
data.drop(data[data['Age'] > 90].index, inplace = True) 
```

```python
sns.boxplot(data['Age']);
```

```python
data.shape
```

The removal of 7 datapoints has given us a much more reasonable spread of data, with a median age of just over 30 and very few people under 20 or over 50.  This is reasonable considering expected ages for the workforce. 

```python
data.Gender.value_counts()
```

There are a multitude of response types for gender, many of which could be combined.  Note the imbalance in the data with many more males.  Let's logically reduce the number of catergories here. 

```python
data.Gender=[c.lower() for c in data.Gender]
data.Gender = data.Gender.replace(to_replace='m', value='male')
data.Gender = data.Gender.replace(to_replace='f', value='female')
data.Gender = data.Gender.replace(to_replace='woman', value='female')
data.Gender = data.Gender.replace(to_replace='make', value='male')
data.Gender = data.Gender.replace(to_replace='msle', value='male')
data.Gender = data.Gender.replace(to_replace='man', value='male')
data.Gender = data.Gender.replace(to_replace='mail', value='male')
data.Gender = data.Gender.replace(to_replace='malr', value='male')
data.Gender = data.Gender.replace(to_replace='femail', value='female')
data.Gender = data.Gender.replace(to_replace='maile', value='male')
data.Gender = data.Gender.replace(to_replace='mal', value='male')
data.Gender = data.Gender.replace(to_replace='male (cis)', value='male')
data.Gender = data.Gender.replace(to_replace='female (cis)', value='female')
data.Gender = data.Gender.replace(to_replace='cis male', value='male')
data.Gender = data.Gender.replace(to_replace='cis man', value='male')
data.Gender = data.Gender.replace(to_replace='cis female', value='female')
data.Gender = data.Gender.replace(to_replace='femake', value='female')
data.Gender = data.Gender.replace(to_replace='female ', value='female')
data.Gender = data.Gender.replace(to_replace='male ', value='male')
data.Gender = data.Gender.replace(to_replace='cis-female/femme', value='female')
data.Gender = data.Gender.replace(to_replace='female (trans)', value='non-binary')
data.Gender = data.Gender.replace(to_replace='androgyne', value='non-binary')
data.Gender = data.Gender.replace(to_replace='nah', value='non-binary')
data.Gender = data.Gender.replace(to_replace='fluid', value='non-binary')
data.Gender = data.Gender.replace(to_replace='agender', value='non-binary')
data.Gender = data.Gender.replace(to_replace='male leaning androgynous', value='non-binary')
data.Gender = data.Gender.replace(to_replace='queer', value='non-binary')
data.Gender = data.Gender.replace(to_replace='ostensibly male, unsure what that really means', value='non-binary')
data.Gender = data.Gender.replace(to_replace='male-ish', value='non-binary')
data.Gender = data.Gender.replace(to_replace='something kinda male?', value='non-binary')
data.Gender = data.Gender.replace(to_replace='genderqueer', value='non-binary')
data.Gender = data.Gender.replace(to_replace='guy (-ish) ^_^', value='non-binary')
data.Gender = data.Gender.replace(to_replace='trans-female', value='non-binary')
data.Gender = data.Gender.replace(to_replace='enby', value='non-binary')
data.Gender = data.Gender.replace(to_replace='neuter', value='non-binary')
data.Gender = data.Gender.replace(to_replace='queer/she/they', value='non-binary')
data.Gender = data.Gender.replace(to_replace='trans woman', value='non-binary')
```

```python
data.Gender.value_counts()
```

### Dashboard Implementation

```python
data.Country.unique()
```

```python
app = dash.Dash(__name__)
server = app.server

country_names = data.Country.unique()
country_names.sort()

app.layout = html.Div([
    html.Div([dcc.Dropdown(id='country-select', 
                           options=[{'label': i, 'value': i} for i in country_names],
                           value='TOR', style={'width': '140px'})]),
    dcc.Graph('shot-dist-graph', config={'displayModeBar': False})])

@app.callback(
    Output('shot-dist-graph', 'figure'),
    [Input('country-select', 'value')]
)

def update_graph(cntrname = 'United States'):
    import plotly.express as px
    return px.scatter(data[data.Country == cntrname], 
                      x = 'Age', y = 'state', 
                      color = 'treatment')

if __name__ == '__main__':
    app.run_server(debug=False)
```
