# Scraping Data from Webpages

1. [US CDC Disease Control and Prevention](https://chronicdata.cdc.gov/Chronic-Disease-Indicators/U-S-Chronic-Disease-Indicators-CDI-/g4ie-h725)

US CDC Disease Control and Prevention Data

```python
#load packages
import requests
import json
import pandas as pd
```

```python
# set url and get page to handle site contents
url = 'https://chronicdata.cdc.gov/resource/g4ie-h725.json'
data = requests.get(url)
#print(data.json())
```

```python
# convert to dataframe
df = pd.json_normalize(data.json())
# look at it
df.head()
```

```python
df.to_csv('USCDCDiseaseControlPrevention.csv', index = False)
```
