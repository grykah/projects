```python
state = search.by_state('New York')

for entry in state:
    zips.append(entry.zipcode)
    zip_type.append(entry.zipcode_type)
    major_city.append(entry.major_city)
    post_office_city.append(entry.post_office_city)
    common_cities_list.append(entry.common_city_list)
    county.append(entry.county)
    st.append(entry.state)
    lat.append(entry.lat)
    lng.append(entry.lng)
    timezone.append(entry.timezone)
    radius_miles.append(entry.radius_in_miles)
    area_codes.append(entry.area_code_list)
    population.append(entry.population)
    population_density.append(entry.population_density)
    land_area_sqmi.append(entry.land_area_in_sqmi)
    water_area_sqmi.append(entry.water_area_in_sqmi)
    housing_units.append(entry.housing_units)
    occupied_housing_units.append(entry.occupied_housing_units)
    median_home_value.append(entry.median_home_value)
    median_household_income.append(entry.median_household_income)
    bounds_west.append(entry.bounds_west)
    bounds_east.append(entry.bounds_east)
    bounds_north.append(entry.bounds_north)
    bounds_south.append(entry.bounds_south)
    polygon.append(entry.polygon)
    pop_by_year.append(entry.population_by_year)
    pop_by_age.append(entry.population_by_age)
    pop_by_gender.append(entry.population_by_gender)
    pop_by_race.append(entry.population_by_race)
    
df.zip = zips
df.zip_type = zip_type
df.major_city = major_city
df.post_office_city = post_office_city
df.common_cities = common_cities_list
df.county = county
df.state = st
df.lat = lat
df.lng = lng
df.timezone = timezone
df.radius_miles = radius_miles
df.area_codes = area_codes
df.population = population
df.population_density = population_density
df.land_area_sqmi = land_area_sqmi 
df.water_area_sqmi = water_area_sqmi
df.housing_units = housing_units
df.occupied_housing_units = occupied_housing_units
df.median_home_value = median_home_value
df.median_household_income = median_household_income
df.bounds_west = bounds_west
df.bounds_east = bounds_east
df.bounds_north = bounds_north
df.bounds_south = bounds_south
df.polygon = polygon
df.population_by_year = pop_by_year
df.population_by_age = pop_by_age
df.population_by_gender = pop_by_gender
df.population_by_race = pop_by_race


print(df)
```
